## Content based recomendation:

```python
df_content=df_copy[['title','hashtags']]
df_content.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english',max_features=5000) #stopwords agey hi nikal dia, fir bhi dekhte hai.
tfidf_matrix = tfidf.fit_transform(df_content['hashtags'])
tfidf_matrix.shape
```

```python
tfidf_matrix=tfidf_matrix.toarray()
```

```python
tfidf_matrix[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(tfidf_matrix)
similarity
```

```python
tfidf.get_feature_names()
```

```python
def get_recomendation(movies):
    #get index
    idx=df_content[df_content['title'] == movies].index[0]
    scores=sorted(list(enumerate(similarity[idx])), key=lambda x: x[1], reverse=True)
    recom=scores[1:11]
    movie_indices = [i[0] for i in recom]
    
    return df_content['title'].iloc[movie_indices]
```

```python
get_recomendation('avatar')
```

```python
get_recomendation('Avatar')
```
